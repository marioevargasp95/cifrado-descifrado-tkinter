```python
import tkinter as tk
from tkinter import messagebox

def cifrar(numero):
    if not numero.isdigit() or len(numero) != 6:
        return "Debe ser un número de 6 dígitos"
    cifras = [(int(d) + 7) % 10 for d in numero]
    cifras[0], cifras[2] = cifras[2], cifras[0]
    cifras[1], cifras[3] = cifras[3], cifras[1]
    cifras[4], cifras[5] = cifras[5], cifras[4]
    return ''.join(map(str, cifras))

def descifrar(numero):
    if not numero.isdigit() or len(numero) != 6:
        return "Debe ser un número cifrado de 6 dígitos"
    cifras = [int(d) for d in numero]
    cifras[0], cifras[2] = cifras[2], cifras[0]
    cifras[1], cifras[3] = cifras[3], cifras[1]
    cifras[4], cifras[5] = cifras[5], cifras[4]
    descifrado = [(d - 7 + 10) % 10 for d in cifras]
    return ''.join(map(str, descifrado))

class App:
    def __init__(self, root):
        self.root = root
        root.title("Aplicación de Cifrado")
        root.geometry("300x200")
        tk.Label(root, text="Autor: Mario Vargas").pack(pady=10)
        tk.Button(root, text="Cifrar número", command=self.ventana_cifrado).pack(pady=5)
        tk.Button(root, text="Descifrar número", command=self.ventana_descifrado).pack(pady=5)

    def ventana_cifrado(self):
        v = tk.Toplevel()
        v.title("Cifrado")
        tk.Label(v, text="Número de 6 dígitos:").pack(pady=5)
        entrada = tk.Entry(v)
        entrada.pack()
        resultado = tk.Label(v, text="")
        resultado.pack()

        def cifrar_y_mostrar():
            resultado.config(text=cifrar(entrada.get()))
        
        tk.Button(v, text="Cifrar", command=cifrar_y_mostrar).pack(pady=5)

    def ventana_descifrado(self):
        v = tk.Toplevel()
        v.title("Descifrado")
        tk.Label(v, text="Número cifrado:").pack(pady=5)
        entrada = tk.Entry(v)
        entrada.pack()
        resultado = tk.Label(v, text="")
        resultado.pack()

        def descifrar_y_mostrar():
            resultado.config(text=descifrar(entrada.get()))
        
        tk.Button(v, text="Descifrar", command=descifrar_y_mostrar).pack(pady=5)

root = tk.Tk()
app = App(root)
root.mainloop()
```
